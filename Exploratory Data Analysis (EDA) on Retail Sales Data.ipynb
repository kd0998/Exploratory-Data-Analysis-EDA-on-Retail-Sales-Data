# Visualization: Present insights through bar charts, line plots, and heatmaps.

```python
#Visualization of null values using the heatmap

sns.heatmap(df.isnull(),yticklabels=False,cmap="inferno")
```

```python
sns.pairplot(df)
```

```python
sns.scatterplot(df.Age)
```

```python
a = df.Quantity.value_counts()
```

```python
plt.pie(a,wedgeprops=dict(width=0.3),autopct="%.2f%%");
plt.show()
```

```python
c=df.groupby("Gender")["Gender"].count()
```

```python
plt.pie(c,labels=c.index,autopct="%.2f%%")
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.title("Product by Age")
plt.xlabel('Age')
plt.ylabel('Product Category')
plt.hist(df.Age, bins = np.arange(8,60,2), color = 'red', edgecolor = 'green');
```

```python
e=df.groupby("Product Category")["Product Category"].count()
```

```python
plt.pie(e,labels=e.index,autopct="%.2f%%")
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.distplot(df['Age'])
plt.show()
```

```python
sns.violinplot(x="Gender",y="Age",data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.barplot(x='Product Category', y='Age',data=df, orient='v')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.kdeplot(x='Quantity', data=df)
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(df)
plt.xticks(rotation=90)
plt.show()
```
